```
# from google.colab import drive
# drive.mount('/content/drive')
import pandas as pd
from autocorrect import spell
import warnings
import re

warnings.filterwarnings('ignore')

path = "/content/drive/My Drive/datacleaning/dataclean2.csv"
df1 = pd.read_csv(path, encoding="ISO-8859-1", dtype=str)
# Dataset is now stored in a Pandas Dataframe
df = df1
```

```
# Create lower case 
df['lower'] = df['present_designation'].str.lower()

# Manage special characters
df['processed'] = df['lower'].str.replace('[^A-Za-z\s]+', '')
# df['processed'].tolist()
df['processed'] = [' '.join([spell(i) for i in str(x).split()]) for x in df['processed'].tolist()]
```

```
import numpy as np
abbreviations = {
    'avp':'assistant vice president',
    'svp':'senior vice president',
    'use':'use standards and evaluation',
    'ps': 'personal secretary',
    'hr' : 'human resources',
    'r d' : 'research and development',
    'tm' : 'talent management',
    'f s o' : 'field sales officer',
    'hod' : 'head of department',
    's c m' : 'supply chain management',
    'dy' :'deputy',
    'coo' : 'chief operating officer',
    'ha' : 'health associate',
    'asst' : 'assistant',
    'abm' :'agi business management',
    'apl' : 'applied product logistics',
    'ase' :'area sales executive',
    'ca' : 'chartered accountant'
    
}


df['processed'] = [re.sub(r'\b' + '|'.join(abbreviations.keys()) + r'\b', lambda m: abbreviations[m.group(0)], spell) for spell in df['processed']]

# Manage Misspellings
misspellings = {
    'parma' : 'pharma',
    'manger' : 'manager',
    'dept':'department',
    'prizer' : 'pfizer',
    'ra' :'research assistant',
    'qlikview' : 'quick view',
    'annalist' : 'analyst',
    'mgmt' : 'management',
    'assistance' :'assistant',
    'frond' : 'front',
    'herma' :'pharma',
    'representant' : 'representative'
    
    
}

df["processed"] = df['processed'].replace(misspellings, regex=True)



# classification into levels and functions
leadership = ['director', 'owner','president','head','cofounder','cio','vp','svp','ceo','founder','board']
managerial = ['manager','gm','dgm','lead','supervisor','advisor','sm','am','dm','chief','territory']
entry = ['engineer', 'executive','scientist','representative','agent','employee','consultant', 'associate','student','entry','fresher','trainee','intern','specialist','get','analyst','office.*','technician','coordinator','secretary','technologist']
adjustments = ['vice','general','senior','junior','assistant','incharge','sr','in charge','management']
functions = ['doctor','injectable','distribution','developer','microbiology','validation','formulation','clinic.*','medico','logistics','so','mr','receptionist','data entry','fitter','controller','field','dispatch','finance','legal','technical','maintenance''finance','regulatory','medicine.*','safety','facility','stores','nutrition.*','manufacturing','operator',"audit.*",'pharma.*','medical.*','warehouse','pro','packaging','commercial','accountant','talent','tm','quality','qc','qa','cqa','research','compliance','production','admin','chemist','demand','regulatory affairs','hr','ehs','supply','human','marketing','sales','business','shift','operations','diet.*','sourcing','purchase','account.*','medical.*','fso','pso','dsm','rd','microbiologist']

# redefining the levels basis leftover functions and adjustments (sometimes the valus is just executive or sr)
allentry = entry + adjustments + functions
allleadership = leadership + adjustments
allmanagerial = managerial + adjustments

df['leadership'] = df['processed'].str.contains('|'.join(allleadership),na=False)
df['managerial'] = df['processed'].str.contains('|'.join(allmanagerial),na=False)
df['entry'] = df['processed'].str.contains('|'.join(allentry),na=False)


# Checking for titles without classification
df['check'] = df['processed'].loc[(df['leadership'] == False) & (df['managerial'] == False) & (df['entry'] == False)]

print ("The % of data with rubbish is",len(df['check'].dropna())/len(df['check'])*100)
# Remove the last % of items which do not make sense
rubbish  = list(dict.fromkeys(df['check'].str.split(expand=True).stack().tolist()))
df['rubbish'] = df['processed'].str.contains('|'.join(rubbish),na=False)

df['check'] = df['processed'].loc[(df['leadership'] == False) & (df['managerial'] == False) & (df['entry'] == False) & (df['rubbish'] == False) ]

print ("The % of unclean data is",len(df['check'].dropna())/len(df['check'])*100)

# df[np.isfinite(df['check'])]
df.dropna()
# print ("The % nan values are ",len(df['check'])/len(df1))


# df
```
